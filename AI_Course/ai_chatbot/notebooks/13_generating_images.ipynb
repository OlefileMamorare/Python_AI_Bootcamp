```python
from dotenv import load_dotenv, find_dotenv
from openai import OpenAI
import os
import requests
import shutil

load_dotenv(find_dotenv(), override=True)

client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

prompt = 'a photorealistic image of a ginger cat curled up on a windowsill, gazing out at a busting city street slick with rain.'

response = client.images.generate(
    model='dall-e-3',
    prompt=prompt,
    style='vivid',
    size='1024x1024',
    quality='standard',
    n=1
)
print(response)
```

```python
image_url = response.data[0].url
print(image_url)
```

```python
image_resource = requests.get(image_url, stream=True)
print('Status code: ', image_resource.status_code)
```

```python
if image_resource.status_code == 200:
    image_name = 'ginger-cat.png'
    with open(image_name, 'wb') as f:
        shutil.copyfileobj(image_resource.raw, f)
else:
    print('Error accessing the image!')
```
