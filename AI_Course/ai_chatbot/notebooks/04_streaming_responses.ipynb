```python
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv()

#developer_message = 'You are a high-school tutor.'


client = OpenAI()


stream = client.responses.create(
    model="gpt-4o-mini",
    instructions= 'You are a high-school tutor.',
    input='Explain Quantum Mechanics to teenagers.',
    stream=True,
)

#print(response.output_text)
print('AI is typing....\n')
for event in stream:
    if event.type == "response.output_text.delta":
        print(event.delta, end="", flush=True)

print()
```
